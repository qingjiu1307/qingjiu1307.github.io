```python
from mxnet import gluon, image, init, nd, autograd
from mxnet.gluon import data as gdata, loss as gloss, model_zoo,nn, utils \
as gutils
import numpy as np
import sys
```

```python
pretrained_net = model_zoo.vision.resnet18_v2(pretrained=True)
pretrained_net.features, pretrained_net.output
```

```python
net = nn.HybridSequential()
#去掉了最后的全连接层
for layer in pretrained_net.features[: -2]:
    net.add(layer)
```

```python
X = nd.random.uniform(shape=(1, 3, 320, 480))
net(X).shape
```

```python
num_classes = 21
net.add(nn.Conv2D(num_classes, kernel_size=1),
       nn.Conv2DTranspose(num_classes, kernel_size=64, padding=16, strides=32))
```

```python
def bilinear_kernel(in_channels, out_channels, kernel_size):
    factor = (kernel_size + 1) // 2
    if kernel_size % 2 == 1:
        centor = factor - 1
    else:
        centor = factor - 0.5
    og = np.ogrid[:kernel_size, :kernel_size]
    filt = (1 - abs(og[0] - centor) / factor) * (1 - abs(og[1] - centor) / factor)
    weight = np.zeros((in_channels, out_channels, kernel_size, kernel_size), dtype='float32')
    weight[range(in_channels), range(out_channels), :, :] = filt
    return nd.array(weight)
```

```python
net[-1].initialize(init.Constant(bilinear_kernel(num_classes, num_classes,64)))
net[-2].initialize(init=init.Xavier())
```

```python
loss = gloss.SoftmaxCrossEntropyLoss(axis=1)

trainer = gluon.Trainer(net.collect_params(), 'sgd', {'learning_rate':0.1, 
                                                      'wd':1e-3})
```

```python
def _get_batch(batch):
    features, labels = batch
    if labels.dtype != features.dtype:
        labels = labels.astype(features.dtype)
    return (features, labels, features.shape[0])
```

```python
def evaluate_accuracy(test_iter, net):
    acc_sum, n = nd.array([0]), 0
    for batch in test_iter:
        features, labels, _ = _get_batch(batch)
        for X, y in zip(features, labels):
            y = y.astype('float32')
            acc_sum += (net(X).argmax(axis=1) == y).sum()
            n += y.size()
        acc_sum.wait_to_read()
    return acc_sum / n
```

```python
import time
def train(train_iter, test_iter, net, loss, trainer, num_epochs):
    for epoch in range(num_epochs):
        train_l_sum, train_acc_sum, n, m, start = 0.0, 0.0, 0, 0, time.time()
        for i, batch in enumerate(train_iter):
            Xs, ys, batch_size = _get_batch(batch)
            ls = []
            with autograd.record():
                y_hats = [net(nd.array([X.asnumpy()])) for X in Xs]
                print(nd.array(y_hats).shape)
                ls = [loss(y_hat, y) for y_hat, y in (y_hats, ys)]
            for l in ls:
                l.backwards()
            trainer.step(batch)
            train_l_sum += sum([l.sum().asscalar() for l in ls])
            n += sum([l.size for l in ls])
            train_acc_sum += sum([(y_hat.argmax(axis=1) == y).sum().asscalar()
                                 for y_hat, y in zip(y_hats, ys)])
            m += sum([y.size for y in ys])
        test_acc = evaluate_accuracy(test_iter, net)
        print('epoch %d, loss %.4f, train acc %.3f, test acc %.3f, '
              'time %.1f sec'
              % (epoch + 1, train_l_sum / n, train_acc_sum / m, test_acc,
                 time.time() - start))
```

```python
train(train_iter, test_iter, net, loss, trainer, 5)
```
