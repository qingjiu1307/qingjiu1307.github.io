# 图片增强
通过一系列随机变化生成大量新的样本，从而减低过拟合的可能
### 常用增强方法：

```python
import matplotlib.pyplot as plt
from mxnet import image
img = image.imdecode(open('C:/Users/Administrator/Desktop/闲的没事换换壁纸玩吧/20171104212141_xUV52.jpeg','rb').read())
plt.imshow(img.asnumpy()) 
plt.show()
```

定义一个辅助函数，给定输入图片上img的增强方法augmentation，将运行多次并画出结果

```python
def apply(img, aug, n=3):
    _, figs = plt.subplots(n, n, figsize=(8,8))
    for i in range(n):
        for j in range(n):
            #转换成float,且这里存在一个copy操作
            x = img.astype('float32')
            #保证输入是合法值
            y = aug(x).clip(0,254)
            figs[i][j].imshow(y.asnumpy()/255.0)
            figs[i][j].axes.get_xaxis().set_visible(False)
            figs[i][j].axes.get_yaxis().set_visible(False)
    plt.show()
```

## 变形
水平方向翻转图片

```python
#以0.5的概率翻转图片
aug = image.HorizontalFlipAug(0.5)
apply(img, aug)
```

随机截取图片中的一块，如果原始输入图片过小，导致没有太多空间随机裁剪，通常先将其放大到足够大的尺寸后再裁剪。

```python
aug = image.RandomCropAug([200, 200])
apply(img, aug)
```

随机裁剪一块随机大小的区域

```python
#随机裁剪，要求至少保留0.1的区域，随机长比在0.5和2之间
#结果再resize到200x200
#物体的大小形状将会发生改变
aug = image.RandomSizedCropAug((200,200), 0.1, (0.5, 2))
apply(img,aug)
```

## 颜色变化

```python
#随机将亮度增加或减小在0 - 50%间的一个量
aug = image.BrightnessJitterAug(0.5)
apply(img, aug)
```

```python
#随机色调的变化
aug = image.HueJitterAug(0.5)
apply(img, aug)
```

## 如何使用
### 数据读取
定义一个辅助函数对图片按顺序应用数个增强

```python
def apply_aug_list(img, augs):
    for f in augs:
        img = f(img)
    return img
```

使用的数据集是 CIFAR10图片，尺寸是32 x 32 x 3, 裁剪成为28 x 28 x 3

```python
train_augs = {
    image.HorizontalFlipAug(0.5),
    image.RandomCropAug((28,28))
}
test_augs = {
    image.CenterCropAug((28, 28))
}
```

定义数据读取，在transform中加入了图片增强

```python
from mxnet import gluon,nd
def get_transform(augs):
    def transform(data, label):
        data = data.astype('float32')
        if augs is not None:
            data = apply_aug_list(data, augs)
        #将data中的数据顺序改成（2，0，1） RGB channel x height x width
        data = nd.transpose(data, (2,0,1))/255
        return data, label.astype('float32')
    return transform
```

```python
def get_data(batch_size, train_augs, test_augs = None):
    cifar10_train = gluon.data.vision.CIFAR10(train=True, transform=get_transform(train_augs))
    cifar10_test = gluon.data.vision.CIFAR10(train=True, transform=get_transform(test_augs))
    train_data = gluon.data.DataLoader(cifar10_train, batch_size, shuffle=True)
    test_data = gluon.data.DataLoader(cifar10_test, batch_size, shuffle=False)
    return (train_data, test_data)
```
