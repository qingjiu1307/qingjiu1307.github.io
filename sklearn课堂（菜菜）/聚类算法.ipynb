```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
x, y = make_blobs(n_samples=500, n_features=2, centers=4, random_state=1)
```

```python
x.shape
```

```python
#对点画图
fig, ax = plt.subplots(1)
ax.scatter(x[:, 0], x[:, 1]
          ,marker='o' #点的形状
          ,s=8 #点的大小
          )
plt.show()
```

```python
colors = ['red', 'pink', 'orange', 'gray']
fig, ax = plt.subplots(1)
for i in range(4):
    ax.scatter(x[y==i, 0], x[y==i, 1]
              ,marker='o'
              ,s=8
              ,c=colors[i])
```

```python
n_clusters = 3
clusters = KMeans(n_clusters=n_clusters, random_state=0).fit(x)
y_pred = clusters.labels_
```

```python
pre = clusters.fit_predict(x)
```

```python
#查看质心
clusters.cluster_centers_
```

```python
#总距离的平方和
clusters.inertia_
```

```python
fig, ax = plt.subplots(1)
for i in range(n_clusters):
    ax.scatter(x[y_pred==i, 0], x[y_pred==i, 1]
              ,marker='o'
              ,s=8
              ,c=colors[i])
centroid = clusters.cluster_centers_
ax.scatter(centroid[:, 0], centroid[:, 1]
          , marker='o'
          ,s=8
          ,c='black')
plt.show()
```

```python
inertials = []
for i in range(3, 7):
    cluster = KMeans(n_clusters=i, random_state=0).fit(x)
    inertials.append(cluster.inertia_)
plt.plot(range(3, 7), inertials)    
```

```python
#轮廓系数
from sklearn.metrics import silhouette_score #所有均值
from sklearn.metrics import silhouette_samples#单个样本
```

```python
silhouette_score(x, y_pred)
```

```python
from sklearn.metrics import calinski_harabaz_score
calinski_harabaz_score(x, y_pred)
```
