```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import svm
from sklearn.datasets import make_circles, make_blobs, make_classification, make_moons
```

```python
n_samples = 100
datasets = [
    make_moons(n_samples=n_samples, noise=.2, random_state=0)
    ,make_circles(n_samples=n_samples, noise=.2, factor=.5, random_state=1)
    ,make_blobs(n_samples=n_samples, centers=2, random_state=5)
    #n_informative  n_redundant
    ,make_classification(n_samples=n_samples, n_features=2, n_informative=2, n_redundant=0, random_state=5)
]
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
for x, y in datasets:
    plt.figure(figsize=(5, 4))
    plt.scatter(x[:, 0], x[:, 1], c=y, s=50,cmap="rainbow")
```

```python
nrows = len(datasets)
ncols = len(kernel) + 1

fig, ax = plt.subplots(nrows, ncols, figsize=(20, 16))
```

```python
#enumerate list[*enumerate] 
[*enumerate(datasets)]
```

```python
#x, y
for ds_cnt, (x,y) in enumerate(datasets):
    #
    a = ax[ds_cnt, 0]
    if ds_cnt == 0:
        a.set_title("Input data")       #zorder
    a.scatter(x[:, 0], x[:, 1], c=y, zorder=10, cmap=plt.cm.Paired, edgecolors='k')
    a.set_xticks(())
    a.set_yticks(())
    #
    for est_idx, kerner in enumerate(kernel):
        
        a = ax[ds_cnt, est_idx+1]
        clf = svm.SVC(kernel=kerner, gamma=2).fit(x, y)
        score = clf.score(x, y)
        
        a.scatter(x[:, 0], x[:, 1], c=y, zorder=10, cmap=plt.cm.Paired
                 , edgecolors='k')
        
        #,facecolor
        a.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1]
                 , s=50, facecolors='none', zorder=10, edgecolors='k')
        #
        xmin, xmax = x[:, 0].min() - .5, x[:, 0].max() + .5
        ymin, ymax = x[:, 1].min() - .5, x[:, 1].max() + .5
        #SVM []
        xx, yy = np.mgrid[xmin:xmax:200j, ymin:ymax:200j]
        #numpy.vstack
        z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
        #
        a.pcolormesh(xx, yy, z>0, cmap=plt.cm.Paired)
        #
        a.contour(xx, yy, z, colors=['k', 'k', 'k'], linestyles=["--", "-","--"]
                  , levels=[-1, 0, 1])
        a.set_xticks(())
        a.set_yticks(())
             # 
        a.text(.95, .06, ('%.2f ' % score).lstrip('0')
               ,size=15
              , bbox=dict(boxstyle='round', alpha=.8, facecolor='white')
              , transform=a.transAxes
              , horizontalalignment='right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from time import time
import datetime
```

```python
data = load_breast_cancer()
x = data.data
y = data.target
```

```python
np.unique(y)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
```

```python
from sklearn.decomposition import PCA
```

```python
X_trans = PCA(2).fit_transform(x)
plt.scatter(X_trans[:, 0], X_trans[:, 1], c=y)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.3, random_state=420)
```

```python
for ker in kernel:
    time0 = time()
    clf = SVC(kernel=ker
             , gamma='auto'
             , degree=1 #1
             , cache_size=5000).fit(xtrain, ytrain)#
    print("the accuracy under kerner %s is %f" % (ker, clf.score(xtest, ytest)))
    #print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))
    print(time() - time0)
```
